#### Dealing with Numerical Data

```python
numerical_columns = data.select_dtypes(include='float').columns
numerical_columns
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(data[numerical_columns])
```

```python
scaled_columns = scaler.transform(data[numerical_columns])
```

```python
scaled_data = pd.DataFrame(scaled_columns,columns=numerical_columns)
```

```python
data = pd.concat([data.drop(numerical_columns,axis=1),scaled_data],axis=1)
```

```python
data.head()
```

transformed data with the standard scaler for better results

### Splitting Data into Training and Testing Sets

```python
X = data.drop(target,axis=1)
y = data[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21)
```

### Model Training

```python
logistic_model = LogisticRegression()
decision_tree_model = DecisionTreeClassifier()
xgb_model = XGBClassifier()
knn_model = KNeighborsClassifier()
random_forest_model = RandomForestClassifier()
```

crreating instances of all the models

```python
logistic_param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
}
```

```python
decision_tree_param_grid = {
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
```

```python
xgb_param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5],
    'subsample': [0.8, 1.0]
}
```

```python
knn_param_grid = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance']
}
```

```python
random_forest_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 20],
    'min_samples_split': [2, 5]
}
```

defining parameter grids for all models

```python
logistic_grid_search = GridSearchCV(logistic_model, logistic_param_grid, verbose=3, scoring='accuracy')
decision_tree_grid_search = GridSearchCV(decision_tree_model, decision_tree_param_grid, verbose=3, scoring='accuracy')
xgb_grid_search = GridSearchCV(xgb_model, xgb_param_grid, verbose=3, scoring='accuracy')
knn_grid_search = GridSearchCV(knn_model, knn_param_grid, verbose=3, scoring='accuracy')
random_forest_grid_search = GridSearchCV(random_forest_model, random_forest_param_grid, verbose=3, scoring='accuracy')
```

```python
logistic_grid_search.fit(X_train, y_train)
decision_tree_grid_search.fit(X_train, y_train)
xgb_grid_search.fit(X_train, y_train)
knn_grid_search.fit(X_train, y_train)
random_forest_grid_search.fit(X_train, y_train)
```

```python
print(logistic_grid_search.best_params_)
print(decision_tree_grid_search.best_params_)
print(xgb_grid_search.best_params_)
print(knn_grid_search.best_params_)
print(random_forest_grid_search.best_params_)
```

```python
pred_log = logistic_grid_search.predict(X_test)
pred_dtree = decision_tree_grid_search.predict(X_test)
pred_xgb = xgb_grid_search.predict(X_test)
pred_knn = knn_grid_search.predict(X_test)
pred_forest = random_forest_grid_search.predict(X_test)
```

```python
print(accuracy_score(y_test,pred_log))
print(accuracy_score(y_test,pred_dtree))
print(accuracy_score(y_test,pred_xgb))
print(accuracy_score(y_test,pred_knn))
print(accuracy_score(y_test,pred_forest))
```
