### Data Preprocessing

```python
data.select_dtypes(include='object').columns
```

```python
data[target].value_counts()
```

```python
data[target] = data[target].map({'No':0,'Yes':1})
data[target].value_counts()
```

```python
data['General_Health'].value_counts()
```

```python
data['General_Health'] = data['General_Health'].map({'Very Good':4,'Good':3,'Excellent':2,'Fair':1,'Poor':0})
data['General_Health'].value_counts()
```
